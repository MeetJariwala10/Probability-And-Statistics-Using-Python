A large circular dartboard is set up with a bulls eye at the center of the circle, which is at coordinate (0,0). A dart is thrown at the center but lands at (X, Y) where X and Y are two different Gaussian Random variables. Write a program to find the average distance of the dart  from the bulls eye for 5, 10, 100, 1000 iterations. Plot the scatter diagram for the same.

```python
import random
import matplotlib.pyplot as plt

i = 0
r = []
x1 = []
y1 = []
avg_list = []
iterations = [5, 10, 100, 1000]

while i < 5:
  i += 1
  x = random.gauss(0, 1)
  x1.append(x)
  y = random.gauss(0, 1)
  y1.append(y)
  d = (x**2 + y**2)**0.5
  r.append(d)

avg1 = sum(r)/i
avg_list.append(avg1)
print("Distance of several points from Bullseye : ")
print("Average distance of the darts from Bullseye : ")
print("For 5 Iterations : ", avg1)
plt.scatter(x1, y1)
plt.scatter(0, 0, color = 'r')
plt.show()

while i < 10:
  i += 1
  x = random.gauss(0, 1)
  x1.append(x)
  y = random.gauss(0, 1)
  y1.append(y)
  d = (x**2 + y**2)**0.5
  r.append(d)

avg2 = sum(r)/i
avg_list.append(avg2)
print(" Distance of several points from Bullseye : ")
print("Average distance of the darts from Bullseye : ")
print("For 10 Iterations : ", avg2)
plt.scatter(x1, y1)
plt.scatter(0, 0, color = 'r')
plt.show()

while i < 100:
  i += 1
  x = random.gauss(0, 1)
  x1.append(x)
  y = random.gauss(0, 1)
  y1.append(y)
  d = (x**2 + y**2)**0.5
  r.append(d)

avg3 = sum(r)/i
avg_list.append(avg3)
print(" Distance of several points from Bullseye : ")
print("Average distance of the dart from Bullseye : ")
print("For 100 Iterations : ", avg3)
plt.scatter(x1, y1)
plt.scatter(0, 0, color = 'r')
plt.show()

while i < 1000:
  i += 1
  x = random.gauss(0, 1)
  x1.append(x)
  y = random.gauss(0, 1)
  y1.append(y)
  d = (x**2 + y**2)**0.5
  r.append(d)

avg4 = sum(r)/i
avg_list.append(avg4)
print(" Distance of several points from Bullseye : ")
print("Average distance of the dart from Bullseye : ")
print("For 1000 Iterations : ", avg4)
plt.scatter(x1, y1)
plt.scatter(0, 0, color = 'r')
plt.show()

plt.title("Average Iterations x Iterations")
plt.plot(iterations, avg_list)
plt.show()
```
