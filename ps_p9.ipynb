```python
# (A) Simulate a fair coin tossing experiment. Using 1000 repetitions of the experiment, count the number of times heads occur.  Comment on obtaining simulation result for n=10, 100 and 10000.

# Random numbers i.e. 0 or 1
# 1 -> Heads
# 0 -> Tails

import random

def coinToss():
    return random.randint(0, 1)

recordList_heads = []
recordList_tails = []

# For n repitions

n = int(input("Enter number of iterations : "))

for i in range(n):
  flip = coinToss()
  if (flip == 0):
    recordList_heads.append(0)
  if(flip == 1):
    recordList_tails.append(1)

print("Heads : ", str(recordList_tails.count(1)))
print("Tails : ", str(recordList_heads.count(0)))
```

```python
# Factorial
def factorial(num):
  fact = 1
  if(num == 0 or num == 1):
    return fact
  for i in range(num):
    fact = fact * (i+1)
  return fact

n_toss = int(input("Enter number of tosses of coin : "))
possibility = 2**(n_toss)
print("Total possibilities : ", possibility)

n = int(input("Enter number of heads : "))

# Using nCr = n! / r! * (n-r)!
nCr = factorial(n_toss)/(factorial(n) * factorial(n_toss - n))
print("probability of obtaining ", n, " heads in ", n_toss, " coin tosses: ", (nCr / possibility), "\n")

# Using Binomial Distribution
print("BY BINOMIAL DISTRIBUTION")
n = int(input("Enter number of tosses of coin : "))
x = int(input("Enter number of heads :"))
p = 1/2
q = 1 - p

binomial_dist = (factorial(n) / (factorial(x) * factorial(n - x))) * (p**x) * (q**(n-x))
print("probability of obtaining ", x, " heads in ", n, " coin tosses: ", binomial_dist, "\n")

```
