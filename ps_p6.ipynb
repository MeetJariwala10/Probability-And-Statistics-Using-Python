The population (p) of a small community on the outskirts of a city grows rapidly over a 20 years period:

t : 0  5  10  15  20

p : 100 200 450 950 2000

As an engineer working for a utility company, you must forecast the population 5 years into the future in order to anticipate the demand for power.

(1) Employ a linear regression model using least square approximation method.

(2) Fit the polynomial function of the 2nd order, 3rd order and 4th order that best fits the data.
                     
(3) Calculate and plot the residuals. (Residual=y_actual - y_approximate (obtained using linear model)
                     
(4) Predict  the population of the community after 25 years in both of the cases.

```python
import numpy as np
import matplotlib.pyplot as plt

# Q-1 --> Employ a linear regression model using least square approximation method
t = np.array([0, 5, 10, 15, 20])
p = np.array([100, 200, 450, 950, 2000])
n = np.size(t)

print("Years : ", t, "\n")
print("Sum of t : ", sum(t), "\n")

sq_t = t**2
print("Square of t : ", sq_t, "\n")
print("Sum of Square of t : ", sum(sq_t), "\n")

print("Let the equation be :  y = a*x + b \n")

print("Population : ", p, "\n")
print("Sum of Population : ", sum(p), "\n")

print("years x population : ", t*p, "\n")
print("sum of (years x population) : ", sum(t*p), "\n")

a = ((sum(t) * sum(p)) - (n * sum(t*p))) / ((sum(t))**2 - (n * sum(sq_t)))
b = ((sum(t) * sum(t*p)) - (sum(p) * sum(sq_t))) / ((sum(t))**2 - (n * sum(sq_t)))

print("Coefficient a : ", a, "\n")
print("Coefficient b : ", b, "\n")

print("Linear equation is : y = ", a, "x + ", b, "\n")

# Q-2 --> Fit the polynomial function of the 2nd order, 3rd order and 4th order that best fits the data.

# Polynomial function of 2nd order
f2 = np.polyfit(t, p, 2)
a2 = round(f2[0], 2)
b2 = round(f2[1], 2)
c2 = round(f2[2], 2)
print("2nd Order Equation :  y = ",a2,"x^2","+",b2,"x +", c2, "\n")

# Polynomial function of 3rd order
f3 = np.polyfit(t, p, 3)
a3 = round(f3[0], 2)
b3 = round(f3[1], 2)
c3 = round(f3[2], 2)
d3 = round(f3[3], 2)
print("3rd Orde Equation :  y = ",a3,"x^3","+",b3,"x^2 +", c3, "x +", d3, "\n")

# Polynomial function of 4th order
f4 = np.polyfit(t, p, 4)
a4 = round(f4[0], 2)
b4 = round(f4[1], 2)
c4 = round(f4[2], 2)
d4 = round(f4[3], 2)
e4 = round(f4[4], 2)
print("4th Order Equation :  y = ",a4,"x^4","+",b4,"x^3 +", c4, "x^2 +", d4, "x +", e4, "\n")

# Q-3 --> Calculate and plot the residuals[residual = actual_y - approx_y (Obtained using linear model)]
y5 =  91.0 * t -170.0
residual = p - y5
print("Residual  : ", residual)

plt.xlabel("Time")
plt.ylabel("Residual")
plt.plot(t, residual)
plt.show()

# Q-4 --> Predict the population of the community after 25 years in both of the cases.
#  x -> Years = 25
#  y -> Population

x = 25

print("Population of the community after 25 years : \n")

y1 = a * x + b
print("For Linear Equation : ", y1)

y2 = a2 * x**2 + b2 * x + c2
print("For 2nd Order Equation : ", y2)

y3 = a3 * x**3 + b3 * x**2 + c3 * x + d3
print("For 3rd Order Equation : ", y3)

y4 = a4 * x**4 + b4 * x**3 + c4 * x**2 + d4 * x + e4
print("For 4th Order Equation : ", y4)
```
