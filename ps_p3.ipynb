Two variables might have same mean, but they might be very different. Look at these two variables.

Profit details of two companies A & B for last 14 Quarters in Millions
1) Company A : 43, 44, 0, 25, 20, 35, -8, 13, -10, -8, 32, 11, -8, 21
2) Company B : 17, 15, 12, 17, 15, 18, 12, 15, 12, 13, 18, 18, 14, 14

Though the average profit is 15 in both the cases. Company B has performed consistently than company A. There was even losses for company A. Measures of dispersion become very vital in such cases.

Calculate the measures of Dispersion - Range,
Inter-Quartile Range, Quartile deviation, Standard Deviation, Variance and coefficient of variation.


```python
import numpy as np
import pandas as pd
from scipy import stats

# Profit Details for Company A & B
m = np.array([43, 44, 0, 25, 20, 35, -8, 13, -10, -8, 32, 11, -8, 21], dtype = int)
n = np.array([17, 15, 12, 17, 15, 18, 12, 15, 12, 13, 18, 18, 14, 14], dtype = int)
print("\nCompany A :", m)
print("Company B :", n)

# Sorting array
a1 = sorted(m)
print("\nCompany A :", a1)

a2 = sorted(n)
print("Company B", a2)

# Range
print("\nRange of A : ", max(a1) - min(a1))
print("Range of B : ", max(a2) - min(a2))



# Inter-Quartile Range
print("\nInter Quartile range of A : ", np.quantile(a1, 0.75) - np.quantile(a1, 0.25))
print("Inter Quartile range of B : ", np.quantile(a2, 0.75) - np.quantile(a2, 0.25))

# Inter-Quartile Range ->
IQR = stats.iqr(a1, interpolation = 'midpoint')
print("Inter quartile range : ", IQR)
IQR = stats.iqr(a2, interpolation = 'midpoint')
print("Inter quartile range : ", IQR)


# Quartile Deviation
print("\nQuartile Deviation of A : ", (np.quantile(a1, 0.75) - np.quantile(a1, 0.25))/2)
print("Quartile Deviation of B : ", (np.quantile(a2, 0.75) - np.quantile(a2, 0.25))/2)

# Standard Deviation
sd1 = np.std(a1, ddof = 1)
sd2 = np.std(a2, ddof = 1)
print("\nStandard Deviation of A : ", sd1)
print("Standard Deviation of B : ", sd2)

# Variance
v1 = sd1 ** 2
v2 = sd2 ** 2
print("\nVariance of A : ", v1)
print("Variance of B : ", v2)

# Coefficient of Variation
print("\nCoefficient of variation of A : ", (sd1 / np.mean(a1))*100)
print("Coefficient of variation of B : ", (sd2 / np.mean(a2))*100)


# Calculating Standard Deviation manually
mean = np.mean(a1)
sum = 0
for i in range(len(a1)):
  sd = (a1[i] - mean) ** 2
  sum = sum + sd

print("\nStandard Deviation of A :", (sum / (len(a1) - 1)) ** 0.5)

```
