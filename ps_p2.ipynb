# Q-1)
There are twelve members in a gym class. Some in good physical condition and some in not-so-good physical condition. Following are  the scores of sit-ups they can complete in a minute
4, 5, 6, 6, 12, 12, 14, 15, 15, 15, 24, 25, 33.
Find measures of central tendency(mean, median, mode) ,  Lower quartile,   Upper quartile, 56th Percentile and 80th to analyze the overall performance.

```python
import numpy as np
import statistics as st

a1 = np.array([ 4, 5, 6, 6, 12, 12, 14, 15, 15, 15, 24, 25, 33], dtype = int)
print("Array is ", a1)
print()

# Mean
print("Mean : ", np.mean(a1))
print()

# Median
print("Median : ", np.median(a1))
print()

# Mode
print("Mode : ", st.mode(a1))
print()

# Percentile
print("56th percentile : ", np.percentile(a1, 56))
print("80th percentile : ", np.percentile(a1, 80))
print()

# Quantile
print("Lower quartile  : ", np.quantile(a1, 0.25))
print("Upper quartile  : ", np.quantile(a1, 0.75))
print()

# Function for calculating percentile
a2 = len(a1)
n = int(input("Enter percentile of :"))
a3 = (n*a2)/100
print(a3)
a4 = int(a3)
print(a4)
print("Percentile : ", a1[a4])
```

# Q-2)
Also find the measures of central tendency,

 Lower quartile,   Upper quartile, 36th Percentile and 90th for data of total bill and tip in tips.csv file.

```python
import pandas as pd
import numpy as np
import statistics as st

Data = pd.read_csv('tips.csv')

print("Mean :")
print("Total bill : ", st.mean(Data['total_bill']))
print("Total tip : ", st.mean(Data['tip']))
print()

print("Median :")
print("Total bill : ", st.median(Data['total_bill']))
print("Tip : ", st.median(Data['tip']))
print()

print("Mode :")
print("Total bill : ", st.mode(Data['total_bill']))
print("Tip : ", st.mode(Data['tip']))
print()

print("Lower Quartile :")
print("Total bill : ", np.quantile(Data['total_bill'], 0.25))
print("Tip : ", np.quantile(Data['tip'], 0.25))
print()

print("Upper Quartile :")
print("Total bill : ", np.quantile(Data['total_bill'], 0.75))
print("Tip : ", np.quantile(Data['tip'], 0.75))
print()

print("36th Percentile :")
print("Total bill : ", np.percentile(Data['total_bill'], 36))
print("Tip : ", np.percentile(Data['tip'], 36))
print()

print("90th Percentile :")
print("Total bill : ", np.percentile(Data['total_bill'], 90))
print("Tip : ", np.percentile(Data['tip'], 90))
print()
```
