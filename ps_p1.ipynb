Download the database from https://media.geeksforgeeks.org/wp-content/uploads/tips.csv
Graphical representation of data from the database(tips.csv) in the form of
1) Scatter Plot
2) Line Chart
3) Bar Graph
4) Histogram
5) Pie Chart
6) Horizontal Bar graph

```python
import matplotlib.pyplot as plt
import pandas as pd
Data = pd.read_csv('tips.csv')
plt.scatter(Data['day'], Data['tip'], color = '#800080'	)
plt.title("Scatter Plot")
plt.xlabel("day")
plt.ylabel("tip")
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
Data = pd.read_csv('tips.csv')
plt.plot(Data['day'], color = 'b')
plt.plot(Data['tip'], color = 'r')
plt.title("Line Chart")
plt.xlabel("tip")
plt.ylabel("day")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
Data = pd.read_csv('tips.csv')
plt.plot(Data['day'], Data['tip'])
plt.title("Line Chart")
plt.xlabel("day")
plt.ylabel("tip")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
Data = pd.read_csv('tips.csv')
plt.bar(Data['day'], Data['tip'], color = 'b')
plt.title("Vertical Bar Graph")
plt.xlabel("day")
plt.ylabel("tip")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
Data = pd.read_csv('tips.csv')
plt.bar(Data['size']-0.2, Data['total_bill'], color = 'b', width = 0.5)
plt.bar(Data['size']+0.2 ,Data['tip'], color = 'r', width = 0.5)
plt.title("Vertical Bar graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
Data = pd.read_csv('tips.csv')
plt.barh(Data['day'], Data['tip'], color = 'b')
plt.title("Horizontal Bar Graph")
plt.xlabel("tip")
plt.ylabel("day")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
Data = pd.read_csv('tips.csv')
plt.pie(Data['tip'])
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
Data = pd.read_csv('tips.csv')
plt.hist(Data['total_bill'], color = 'g')
plt.title("Histogram")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
