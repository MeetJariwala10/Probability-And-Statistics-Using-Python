The year wise profit of a company (in lakhs) is given in table below:

Year:     2001     2002    2003    2004     2005     2006     2007     2008     2009     2010     2011

Profit:     4            8           9          7             11         15          10         13          18         16        20

Find (a) 3-years moving average (b)  4- years moving average    (c)   5-years moving average.
Fit a trend line for each case.

```python
import numpy as np
import matplotlib.pyplot as plt

# Years
years = np.array([2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011])
print("Years : ", years, "\n")

# Profit
profit = np.array([4, 8, 9, 7, 11, 15, 10, 13, 18, 16, 20])
print("Profit : ", profit, "\n")

# Moving average for 3 years

# For 3 years
y = 3
i = 0

moving_average3 = []
year3 = []

while i < len(years) - y + 1:
  profit_average = round(np.sum(profit[i : i + y]) / y, 2)
  year_average = round(np.sum(years[i : i + y]) / y)
  moving_average3.append(profit_average)
  year3.append(year_average)
  i += 1

print("3-Years Moving Average :\n")
print(year3)
print(moving_average3)

# Plotting graph for 3-Years Moving Average
plt.title("3-Years Moving Average")
plt.xlabel("Years")
plt.ylabel("Profit")
plt.plot(years, profit, label = "Actual data")
plt.plot(year3, moving_average3, label = "Trend Line", linestyle = '--', marker = 'o')
plt.legend()
plt.grid()
plt.show()
```

```python
# Moving average for 4 years

# For 4 years
y = 4
i = 0

moving_average4 = []
year4 = []

while i < len(years) - y + 1:
  profit_average = round(np.sum(profit[i : i + y]) / y, 2)
  year_average = round(np.sum(years[i : i + y]) / y, 2)
  moving_average4.append(profit_average)
  year4.append(year_average)
  i += 1

i = 0
y = 2

m_avg4 = []
y4 = []

while i < len(year4) - y + 1:
  profit_average = round(np.sum(moving_average4[i : i + y]) / y, 2)
  year_average = round(np.sum(year4[i : i + y]) / y, 2)
  m_avg4.append(profit_average)
  y4.append(year_average)
  i += 1

print("4-Years Moving Average :\n")
print(y4)
print(m_avg4)

# Plotting graph for 3-Years Moving Average
plt.title("4-Years Moving Average")
plt.xlabel("Years")
plt.ylabel("Profit")
plt.plot(years, profit, label = "Actual data")
plt.plot(y4, m_avg4, label = "Trend Line", linestyle = '--', marker = 'o')
plt.legend()
plt.grid()
plt.show()
```
